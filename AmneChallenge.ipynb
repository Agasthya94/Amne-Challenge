```python

"""
Author: Agasthya Vidyanath Rao Peggerla
"""

fname = open('test.txt')
fout = open('output.txt',"w")

# A dictionary to store the window result
rmap = {"ones":0,"c_ones":0,"minusones":0,"c_minusones":0}

num_win = fname.readline().strip().split()
line = fname.readline().strip().split()

n = int(num_win[0])
k = int(num_win[1])

arr = [0]*n
j,flag = 0,0


for i in range(len(line) - 1):
    
    # SLide the window
    
    if((i-j) >= (k-1)):
        win_sum = rmap["ones"] + rmap["c_ones"] - rmap["minusones"] - rmap["c_minusones"]
        
        # Write the output to a file
        
        output = str(win_sum) + "\n"
        fout.write(output)
        
        # Make adjustments for next window
        
        if(arr[j+1] == 1):
            rmap["ones"] = rmap["ones"] - 1
            if(arr[j+1] == arr[j+2]):
                rmap["c_ones"] = rmap["c_ones"] - 1
        
        if(arr[j+1] == -1):
            rmap["minusones"] = rmap["minusones"] - 1
            if(arr[j+1] == arr[j+2]):
                rmap["c_minusones"] = rmap["c_minusones"] - 1
        
        j = j+1
    # Compare with the next element and insert into the dict for computation   
    
    if(line[i+1] > line[i]):
        arr[i+1] = 1
        if(flag == 2):
            rmap["ones"] = rmap["ones"] + 1
        elif(flag == 1):
            rmap["ones"] = rmap["ones"] + 1
            rmap["c_ones"] = rmap["c_ones"] + 1
            flag = 2
                
        else:
            rmap["ones"] = rmap["ones"] + 1
            flag=1
        
    elif(line[i+1] < line[i]):
        arr[i+1] = -1
        if(flag == -2):
            rmap["minusones"] = rmap["minusones"] + 1
            
        elif(flag == -1):
            rmap["minusones"] = rmap["minusones"] + 1
            rmap["c_minusones"] = rmap["c_minusones"] + 1
            flag = -2
                
        else:
            rmap["minusones"] = rmap["minusones"] + 1
            flag = -1
        
    
win_sum = rmap["ones"] + rmap["c_ones"] - rmap["minusones"] - rmap["c_minusones"]
output = str(win_sum)
fout.write(output)

fout.close()
        
        
                
                
                
```
